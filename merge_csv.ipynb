```python
import pandas as pd
df1=pd.read_csv('/content/drive/MyDrive/paragraphs.csv')
df2=pd.read_csv('/content/drive/MyDrive/question_answers.csv')
```

```python
dict1={}
for i in range (0,1179):
  dict1[df1['id'][i]]=df1['paragraph'][i]
```

```python
list1=[]
for i in range (0,944):
  list1.append(dict1[df2['paragraph_id'][i]])

df2['context']=list1
```

```python
df2
```

```python
df2.drop('paragraph_id',axis=1)
```

```python
list2=[]
import numpy as np
for i in range (len(df2)):
  listgen=[]
  listgen.append(df2['context'][i].find(df2['answer'][i]))
  list2.append(listgen)

df2['Answer_start']=list2
df2['index']=np.arange(len(df2))
```

```python
df2
```

```python
# rankings_pd.rename(columns = {'test':'TEST'}, inplace = True)
# Q capital
# T capital
# A capital Answer_text
# Paragraph in place of context

df2.rename(columns= {'question':'Question', 'theme':'Theme','answer':'Answer_text','context':'Paragraph'}, inplace=True)
df2.drop('paragraph_id',axis=1)
```

```python
list3=[]
for i in range(len(df2)):
  if(len(df2['Answer_text'][i])==0):
    list3.append(False)
  else:
    list3.append(True)

df2['Answer_possible']=list3
df2.drop('paragraph_id',axis=1)
```
